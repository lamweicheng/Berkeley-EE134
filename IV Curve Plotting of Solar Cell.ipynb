```python
import numpy as np
import matplotlib.pyplot as plt

IL = 1 
Rs = 0.1  
Rp = 100  
I0 = 10e-6  
T = 300  
q = 1.602e-19  
k = 1.38e-23  
n = 2  


def calculate_current(V):
    return IL - I0 * (np.exp((q * (V + Rs * IL)) / (n * k * T)) - 1) - (V + Rs * IL) / Rp


V_points = np.linspace(-0.75, 0.75, 100)


I_points = [calculate_current(V) for V in V_points]


plt.plot(V_points, I_points, label='I-V Characteristic')

plt.grid(True, linestyle='--', alpha=0.5)
plt.title('Solar Cell I-V Characteristic')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.legend()

plt.xlim([-0.75, 0.75])
plt.ylim([-2, 2])

plt.show()
```
